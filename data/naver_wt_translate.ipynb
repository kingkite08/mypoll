```python
import json
# JSON 파일 읽기
with open('naver_wt_en.json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
import json
from langchain_openai import ChatOpenAI
from langchain.schema import HumanMessage

# OpenAI Chat 모델 초기화
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.3)

# 데이터 처리 함수
def process_data(data, llm):
    """Translate all fields in the dataset based on the provided conditions."""
    for entry in data:
        print(f"{entry['title']} 번역 시작")
        
        # 번역 조건 및 원본 데이터를 포함한 프롬프트 생성
        prompt = f"""
You are a professional translator with expertise in Korean and English. Translate the following JSON object into English while following these rules:
1. Do not translate 'id', 'thumbnail', 'views', 'url', 'episode', 'rating'.
2. All translated letters must be lower alphabet. Except 'description'.
3. Understand and preserve the contextual meaning and nuance of each Korean term. Translate not just the literal meaning, but also the implied, cultural, or emotional context of each word.
4. For the 'keywords' field, ensure that any specific terms related to Korean webtoons are translated based on the context and cultural understanding:
    - "일상": "slice of life"
5. For the 'author', 'illustrator', 'original' field:
    - If the name is a creative or symbolic nickname like "글개미", transliterate it into Roman characters (e.g., "Geulgaemi").
6. For the 'price' field:
    - Translate "무료" as "free".
    - Translate "유료" as "paid".
    - Translate "24시간 기다리면 무료"and similar phrases as: "free after waiting for X [day(s)/hour(s)]".
    - Always use this structure: "free after waiting for [time period]".
    - Ensure consistency in grammar and style.
7. Fields with a value of "-" should remain as "-".
8. Maintain the structure of the JSON object exactly as it is.
9. For the 'status' field:
    - Translate "월요웹툰" as "every monday", "화요웹툰" as "every tuesday" and so on. Similar phrases as "every [day(s)] series".
    - Translate "12화 완결" as "12 episodes completed".
10. For the "age_rating" field:
    - Translate "전체연령가" as "all".
    - Translate "15세 이용가" as "rated 15+".
    - Translate "19세 이용가" as "rated 19+".
11. For the "description" field:
    - It is a synopsis of the webtoon. Remove all the "\n","\r" characters and start translating.
12. For the 'genre' field:
    - "일상": "slice of life"

Here is the JSON object to translate:
{json.dumps(entry, ensure_ascii=False, indent=4)}

Respond only with the translated JSON object.
"""
        try:
        # OpenAI 모델 호출
            response = llm.invoke([HumanMessage(content=prompt)])
            
            # 응답 내용 출력 (디버깅용)
            print("Raw response:", response.content.strip())
            
            # 번역된 JSON 객체를 파싱하여 업데이트
            translated_entry = json.loads(response.content.strip())
            entry.update(translated_entry)
        except Exception as e:
         print(f"Error processing entry with ID {entry.get('id')}: {e}") 
```

```python
process_data(data,llm)
```

```python
with open('translated.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```
